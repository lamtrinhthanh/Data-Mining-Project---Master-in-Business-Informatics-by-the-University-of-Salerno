```python
#QUESTION 1
def statisticalDescription(dataset, *attributes):
    '''To provide the statistical description of one or some given attributes of the dataset'''
    for attribute in attributes:
        stat_metrics = dataset[attribute].describe()
        mode = stats.mode(dataset[attribute])
        print("================================")
        print('Attribute:\033[1m', attribute.upper(), '\033[0;0m')
        print('Number of oberservations:', int(stat_metrics[0]))
        print('Mean:', round(stat_metrics[1], 2))
        print('Median:', round(stat_metrics[5], 2))
        print('Mode:', float(mode[0]), 'with frequency:', float(mode[1]))
        print('Standard deviation:', round(stat_metrics[2], 2))
        print('Minimum value:', round(stat_metrics[3], 2))
        print('Maximum value:', round(stat_metrics[-1], 2))
        print('Lower quartile (25%):', round(stat_metrics[4], 2))
        print('Upper quartile (75%):', round(stat_metrics[6], 2))
    print(("================================"))
    print('\n')
```

```python
#QUESTION 1
def removeOutliers(dataset, *attributes, value=1.5):
    '''Remove the outliers in attribute(s) of a dataset applying 1.5*IQR rule
    value is set to default as 1.5, but it could be set to some other numbers (such as 3)
    Return a processed_dataset'''

    for attribute in attributes:
        Q3 = dataset[attribute].quantile(0.75)
        Q1 = dataset[attribute].quantile(0.25)
        IQR = Q3 - Q1
        upper_outlier = Q3 + value * IQR
        lower_outlier = Q1 - value * IQR
        dataset = dataset[~(dataset[attribute] < lower_outlier)
                          & ~(dataset[attribute] > upper_outlier)]
        return dataset
```

```python
#QUESTION 4
def trend_recipe(dataset,
                    recipe_id,
                    start,
                    end,
                    id_column='recipe_id',
                    date_column='date',
                    rating_column='rating'):
    """Plot the trend for the average rating of a specified recipe id given the period by year
    df: the dataset
    recipe_id: the id in interest
    start: start point of the period
    end: end point of the period
    """
    df = copy.deepcopy(dataset)
    if int(start) == int(end):
        end = str(int(end) + 1)
    df[date_column] = df[date_column].dt.year
    filtered_df = df[(df[id_column] == recipe_id)
                     & (df[date_column] >= int(start)) &
                     (df[date_column] <= int(end))].groupby(
                         date_column)[rating_column].mean().round(2)
  
    if len(filtered_df) > 1:
        plt.figure(figsize=(10, 5))
        plt.plot(filtered_df.index, filtered_df.values)
        plt.xticks(filtered_df.index)
        plt.xlabel('Time')
        plt.ylabel('Average Rating')
    else:
        plt.figure(figsize=(10, 5))
        plt.scatter(filtered_df.index, filtered_df.values)
        plt.xticks(filtered_df.index)
        plt.xlabel('Time')
        plt.ylabel('Average Rating')
    plt.show()
```

```python
#QUESTION 4
def trend_all_recipes(dataset,
                    start,
                    end,
                    mode,
                    id_column='recipe_id',
                    date_column='date',
                    rating_column='rating'):
    """Plot the trend for the average rating of a specified recipe id given the period by year
    df: the dataset
    start: start point of the period
    end: end point of the period
    mode: sort by given unit ('D' - days, 'M' - months, 'Y' - years)
    """
    filtered_df = dataset[(dataset[date_column] >= start) & (dataset[date_column] <= end)]
    grouped_df = filtered_df.groupby(pd.Grouper(
    key=date_column, freq=mode))[rating_column].mean()
    plt.figure(figsize = (10,10))
    plt.plot(grouped_df.index, grouped_df.values)
    plt.xlabel('Time')
    plt.ylabel('Average Rating')
    plt.show()    
```

```python
#QUESTION 5
def submission_trend(dataset, mode, start = '1999', end = '2019', submission_column='submitted'):
    """The function help plots the trend of recipes submitted over the given period of time
    dataset: the data from which to plot the trend
    mode: {y: year, m: month, d: day}
    start: starting time for the trend plotting
    end: end time for the trend plotting"""
    #create a copy of the original dataset, convert the date attributes into datetime format and filter
    df = copy.deepcopy(dataset)
    df[submission_column] = pd.to_datetime(df[submission_column])
    df = df[(df[submission_column] >= start) & (df[submission_column] <= end)]
    
    #apply different mode 'd', 'm', 'y' for plotting, x tick: the labels for x-axis
    if (mode.lower() == 'y') or (mode.lower() == 'year'):
        submitted_trend = df.groupby(pd.Grouper(key = submission_column, freq = 'y', label = 'right')).size().reset_index()  
        x_tick = submitted_trend[submission_column].dt.year
        x_label = 'year'
    elif (mode.lower() == 'm') or (mode.lower() == 'month'):
        submitted_trend = df.groupby(pd.Grouper(key = submission_column, freq = 'm', label = 'right')).size().reset_index()
        x_tick = submitted_trend[submission_column].dt.to_period('m').apply(lambda x: str(x))
        x_label = 'month'
    elif (mode.lower() == 'd') or (mode.lower() == 'date'):
        submitted_trend = df.groupby(pd.Grouper(key = submission_column, freq = 'd', label = 'left')).size().reset_index()
        x_tick = submitted_trend[submission_column]
        x_label = 'day'
    
    #make a histogram plot 
    submitted_trend[submission_column] = x_tick
    fig = plt.figure(figsize= (len(submitted_trend[submission_column]), 10))
    ax = plt.gca()
    plt.bar(x_tick, submitted_trend.iloc[:, -1], width =1, edgecolor='black')
    ax.set_xticks(x_tick)
    plt.xticks(rotation = 'vertical', fontsize = 15)
    ax.set_xlabel('Trend plotted per ' + x_label + ' from ' + start + ' to ' + end, fontsize= 15)
    ax.xaxis.labelpad = 20
    ax.set_ylabel('The number of submissons', fontsize= 15)
    ax.yaxis.labelpad = 20
    plt.yticks(fontsize = 15)
    plt.show()
```

```python
#Question 7, 8, 9, 10
def text_preprocessing(text):
    """text: input text to be pre-processed by reomoving unnecessary characters, space,...
    return: a list of words in the text"""
    pattern = r'[\'\"\[\]]'
    return sorted(word.strip() for word in re.sub(pattern, '', text).strip().split(','))
```

```python
def text_preprocessing_2(text):
    """text: input text to be pre-processed by reomoving unnecessary characters, space,...
    return: a processed string"""
    pattern = r'[\'\"\[\]]'
    return re.sub(pattern, '', text).strip()
```

66th quantile of averaged ratings by recipe ids is 5.
This shows that most of the values > 50 of the values is 5, so it is difficult to ensure the equal distribution of data into categories.

So many 5 means so many excellent recipes. 

In this case, we can try to minimize the bias of our dataset by choosing a percentile value as the maximum value for "Good" category but less than 5

```python
ordinal_rating['averaged_rating'].quantile(0.6)
```

```python
ordinal_rating['averaged_rating'].quantile(0.56)
```

```python
ordinal_rating['ordinal_value'] = pd.cut(ordinal_rating['averaged_rating'],
                                         bins=[0, poor_max, good_max, excellent_max],
                                         labels=['poor', 'good', 'excellent'],
                                         include_lowest=True)
```

```python
recipe_map = {}  # key as recipe id and values as tags
recipe_min_hash_map = {}  # store every computed minHash for later use
threshold = 0.7  #similarity threshold
num_perm = 128  #number of permutations or hash functions applied
lsh = MinHashLSH(threshold=threshold, num_perm=num_perm)
```

Scan through each record or observation in **lsh_dataset**, retrieve id and tags, add the corresponding data to **recipe_map =({id: tag}**, **recipe_min_hash_map = {id: minhashed signatures}**, populate **(id, minhashed signatues)** into **lsh function**)

We randomly scan through tags attribute of 20 records to see how text data is stored and formatted in advance of pre-processing text data for minhash and lsh

```python
for i in range(20):
    field_value = LSH_recipes_tags.loc[np.random.randint(0, len(LSH_recipes_tags)), 'tags']    
    print(field_value)
    print('Type:', type(field_value), end = '\n' + '-'*100 + ' \n')
```

We can see that the values of the attribute ***tags*** are stored as text (type: str), including unncessary symbols such as: ',[,], white space, so we only need to retrieve the main contents of each value such as: 'time-to-make', 'course'..., and remove other symbols. Finally, we put them into a list which is used for the minhash and LSH

```python
for row in LSH_recipes_tags.itertuples():
    ids = row.id
    tags = sorted(re.sub(r'[\'\"\[\] ]', '', row.tags).split(',')) #remove unneccessary symbols, 
                                                           #use list to store the tags for each record
    if ids in recipe_map:
        continue 
    recipe_map[ids] = tags
    min_hash = MinHash(num_perm = num_perm)  #assign MinHash function to min_hash with num_perm
    for tag in tags: #
        min_hash.update(tag.encode('utf-8')) # let's populate each tag in tags into minhassh function
    lsh.insert(ids, min_hash) #populate minhash signatures for each id into lsh
    recipe_min_hash_map[ids] = min_hash    
```

The column attribute ***nutritions*** will be the data for training K-Means clustering model, with each row representing 1 example or recipe in our dataset

```python
#preprocessing, transform the values for each nutrition cell into a numerical list
train_data = nutrition_df['nutrition'].apply(text_preprocessing).apply(lambda x: list(map(lambda i: float(i), x))).to_list()
train_data = np.array(train_data)
train_data
```

```python
print('Mean of each feature column in the array:\n', str(np.mean(train_data, axis = 0)))
print('Standard deviation of each feature column in the array:\n', str(np.std(train_data, axis = 0)))
```

Remove outliers and noise

```python
#define the first and third quaters for the feature columns
thredhold_parameter = 1.5
Q1_vector = np.quantile(train_data, q=0.25, axis=0)
Q3_vector = np.quantile(train_data, q=0.75, axis=0)
IQR = Q3_vector - Q1_vector
upper_outlier = Q3_vector + thredhold_parameter * IQR
lower_outlier = Q1_vector - thredhold_parameter * IQR

processed_train_data = train_data[np.all(np.logical_not(
    np.logical_or(train_data < lower_outlier, train_data > upper_outlier)),
                                            axis=1), :]
```

```python
print(
    '%d examples have been removed due to containing outliers or noise values'
    % (len(train_data) - len(processed_train_data)))
```

Standardizing train data

```python
scaler = StandardScaler()
processed_train_data = scaler.fit_transform(processed_train_data)
display(processed_train_data)
```

Apply KMeans Clustering

We define what is the best number of clusters through the plot of ***number of clusters*** against ***Within Cluster Sum of Squares (WCSS)*** 
(WCSS is defined as the sum of the squared distance between each member of the cluster and its centroid)

```python
wcss = []
for i in range(10, 100, 10):
    kmeans = KMeans(n_clusters=i, max_iter = 300)
    kmeans.fit(processed_train_data)
    wcss.append(kmeans.inertia_)    
```
